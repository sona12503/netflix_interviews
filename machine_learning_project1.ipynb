```python
import pandas as pd
data = pd.read_csv("/content/netflix_titles_nov_2019.csv")
print(data.head())
```

```python
print("Content Types:", data['type'].unique())

```

```python
# Calculate the split between Movies and TV Shows
content_distribution = data['type'].value_counts(normalize=True) * 100
print("\nDistribution (%):\n", content_distribution)
```

```python

# Visualize the split (optional)
import matplotlib.pyplot as plt
content_distribution.plot(kind='bar', color=['#E50914', '#221F1F'])  # Netflix colors!
plt.title("Movies vs. TV Shows on Netflix")
plt.ylabel("Percentage (%)")
plt.show()
```

```python
yearly_trend = data['release_year'].value_counts().sort_index()

# Plot the trend
plt.figure(figsize=(12, 6))
yearly_trend.plot(kind='line', marker='o', color='red')
plt.title("Netflix Content Added Per Year (All Types)")
plt.xlabel("Year")
plt.ylabel("Number of Titles")

plt.annotate("Streaming Launched", xy=(2007, 500), xytext=(1995, 1500),
             arrowprops=dict(arrowstyle="->"), fontsize=10)
plt.annotate("Global Expansion", xy=(2016, 2500), xytext=(2016, 3000),
             arrowprops=dict(arrowstyle="->"))
plt.grid(True)
plt.show()
```

```python
trend_by_type = data.groupby(['release_year', 'type']).size().unstack()

# Plot stacked area chart
trend_by_type.plot(kind='area', figsize=(12, 6), color=['#E50914', '#221F1F'])
plt.title("Movies vs. TV Shows Added Per Year")
plt.xlabel("Year")
plt.ylabel("Number of Titles")
plt.legend(title="Content Type")
plt.show()
```

```python
oldest_shows = (
    data[data['type'] == 'Movie']
    .sort_values('duration')
    .head(1)[['title', 'release_year', 'duration']]
)

# Display with stylish formatting
print("📺 Oldest movies on Netflix:")
display(oldest_shows.style.set_caption("Netflix's Vintage Collection"))
```

```python
interview_keywords = [
    'interview', 'conversation', 'q&a', 'roundtable',
    'talk', 'discussion', 'sitdown', 'bonus'
]

# Create mask for interviews
mask = data['title'].str.lower().str.contains('|'.join(interview_keywords)) | \
       data['description'].str.lower().str.contains('|'.join(interview_keywords))

# Filter and preview
interviews = data[mask][['title', 'type', 'duration', 'release_year']]
print(f"Found {len(interviews)} interview-style pieces:")
display(interviews.sample(3, random_state=42))  # Random sample for reproducibility
```

```python
interviews = data[
    (data['duration'] < '15 min') &
    (data['title'].str.contains('interview|conversation', case=False))
]
print(f"Found {len(interviews)} fake 'movies':")
display(interviews[['title', 'duration']])
```

```python
data['content_type'] = data['type']  # Preserve original
data.loc[mask, 'content_type'] = 'Interview'  # Reclassify
print("\nContent Type Distribution:")
print(data['content_type'].value_counts())
```

```python
data['content_type'].value_counts().plot(
    kind='pie', autopct='%.1f%%',
    colors=['#E50914', '#221F1F', '#B81D24'],  # Netflix palette
    title='Netflix Content Types (After Reclassification)'
)
plt.savefig('netflix_content_types.png', bbox_inches='tight')
```

```python
original_genres = data[data['type'] == 'Movie']['listed_in'].value_counts(normalize=True).head(5)
print("Top 5 'Movie' Genres (Before Cleaning):")
display(original_genres)
```

```python
true_movies = data[
    (data['content_type'] == 'Movie') &
    (~data['title'].str.contains('interview|conversation', case=False))
]

# Corrected genre distribution
corrected_genres = true_movies['listed_in'].value_counts(normalize=True).head(5)
print("\nTop 5 Genres (After Cleaning):")
display(corrected_genres)
```

```python
pd.DataFrame({'Before': original_genres, 'After': corrected_genres}).plot(kind='bar', figsize=(10,6))
plt.title("Genre Popularity: Before vs After Cleaning")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('genre_shifts.png')
```
